### FEATURES
The various features of the cleaned dataset are explained below:
1) Airline: The name of the airline company is stored in the airline column. It is a categorical feature having 6 different airlines.
2) Flight: Flight stores information regarding the plane's flight code. It is a categorical feature.
3) Source City: City from which the flight takes off. It is a categorical feature having 6 unique cities.
4) Departure Time: This is a derived categorical feature obtained created by grouping time periods into bins. It stores information about the departure time and have 6 unique time labels.
5) Stops: A categorical feature with 3 distinct values that stores the number of stops between the source and destination cities.
6) Arrival Time: This is a derived categorical feature created by grouping time intervals into bins. It has six distinct time labels and keeps information about the arrival time.
7) Destination City: City where the flight will land. It is a categorical feature having 6 unique cities.
8) Class: A categorical feature that contains information on seat class; it has two distinct values: Business and Economy.
9) Duration: A continuous feature that displays the overall amount of time it takes to travel between cities in hours.
10)Days Left: This is a derived characteristic that is calculated by subtracting the trip date by the booking date.
11) Price: Target variable stores information of the ticket price.

###  Additional Business Insights to Add with Plots

## 1. Airline Price Comparison
Insight: Identify which airlines are premium and which are budget based on average price.

```python
# Plot: Bar plot of average flight prices per airline.
plt.figure(figsize=(12, 6))
sns.barplot(x='Airline', y='Price', data=df, estimator=np.mean, palette='coolwarm')
plt.xticks(rotation=45)
plt.title('Average Price per Airline')
plt.ylabel('Average Price')
plt.xlabel('Airline')
plt.tight_layout()
plt.show()
```

### 2. Price Trend by Travel Month
Insight: Show how prices vary by month — useful for seasonal demand analysis.

Plot: Line or bar plot of average price by Month.

```python
monthly_price = df.groupby('Month')['Price'].mean().reset_index()
plt.figure(figsize=(10,6))
sns.lineplot(data=monthly_price, x='Month', y='Price', marker='o', color='orange')
plt.title('Average Flight Price per Month')
plt.ylabel('Average Price')
plt.xlabel('Month')
plt.grid(True)
plt.show()
```

### 3. Impact of Stops on Pricing
Insight: Show how the number of stops (direct vs 1-stop vs 2-stop) affects price.

Plot: Boxplot or barplot of Price by Total_Stops.



```python
plt.figure(figsize=(8,6))
sns.boxplot(x='Total_Stops', y='Price', data=df, palette='Set2')
plt.title('Price Distribution by Number of Stops')
plt.ylabel('Price')
plt.xlabel('Total Stops')
plt.show()
```

### 4. Price vs Duration
Insight: Longer flights may not always be more expensive; compare duration with price.

Plot: Scatter plot of Duration (converted to minutes) vs Price.

First, convert Duration to total minutes:




```python
def duration_to_minutes(duration_str):
    h, m = 0, 0
    if 'h' in duration_str:
        h = int(duration_str.split('h')[0].strip())
        duration_str = duration_str.split('h')[1]
    if 'm' in duration_str:
        m = int(duration_str.split('m')[0].strip())
    return h * 60 + m

df['Duration_mins'] = df['Duration'].apply(duration_to_minutes)

plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='Duration_mins', y='Price', hue='Airline', alpha=0.6)
plt.title('Price vs Duration (in minutes)')
plt.xlabel('Duration (minutes)')
plt.ylabel('Price')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

### 5. Cheapest Time to Fly
Insight: Reveal the cheapest hours to depart or arrive.

Plot: Bar plots for average price by Departure_hour and Arrival_hour.




```python
plt.figure(figsize=(12,5))
sns.barplot(data=df, x='Departure_hour', y='Price', estimator=np.mean, palette='Blues')
plt.title('Average Price by Departure Hour')
plt.xlabel('Hour of Day')
plt.ylabel('Average Price')
plt.show()

plt.figure(figsize=(12,5))
sns.barplot(data=df, x='Arrival_hour', y='Price', estimator=np.mean, palette='Greens')
plt.title('Average Price by Arrival Hour')
plt.xlabel('Hour of Day')
plt.ylabel('Average Price')
plt.show()
```
